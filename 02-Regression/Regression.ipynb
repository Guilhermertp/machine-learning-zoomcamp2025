```python
# format the columns to replace spaces with underscore and lowercase
df.columns =df.columns.str.replace(" ","_").str.lower()
```

```python
df.head()
```

```python
# check the type of the columns to find the string columns
df.dtypes
```

```python
string_columns = []

# get all the columns of type object
for i in df.columns:
    if df[i].dtype == 'O':
        #print(df[i].name)
        string_columns.append(df[i].name)
print(string_columns)
```

```python
# another way to get the object columns
strings= list(df.dtypes[df.dtypes=='object'].index)
strings
```

```python
for col in strings:
    df[col] = df[col].str.lower().str.replace(' ','_')
```
